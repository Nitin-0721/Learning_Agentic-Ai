## Models Integration with OpenAi, Gemini, Groq ##

```python
import os
from dotenv import load_dotenv
load_dotenv()

os.environ["OPENAI_API_KEY"]=os.getenv("OPENAI_API_KEY")
os.environ["GOOGLE_API_KEY"]=os.getenv("GOOGLE_API_KEY")
os.environ["GROQ_API_KEY"]=os.getenv("GROQ_API_KEY")
```

```python
from langchain.chat_models import init_chat_model

model = init_chat_model(
    "llama-3.3-70b-versatile",
    model_provider="groq"
)

model
```

```python
##invoke the model
response = model.invoke("Hello how are u?")
response
```

```python
response.content
```

### Groq Model Integration ###

```python
import os
from langchain.chat_models import init_chat_model
os.environ["GROQ_API_KEY"] = os.getenv("GROQ_API_KEY")

model = init_chat_model("groq:qwen/qwen3-32b")
response = model.invoke("Distance between India to New Zealand from all mode of transport?")
response.content
```

## Streaming and Batch ##

```python
##Streaming

for chunk in model.stream("Write 200 words on music"):
    print(chunk.text, end = "|", flush = True)
```

```python
##Batch

responses = model.batch([
    "Why people smoke?",
    "Why girls are attractive?",
    "How to make anyone fool?"
])
for response in responses:
    print(response.content)
```
